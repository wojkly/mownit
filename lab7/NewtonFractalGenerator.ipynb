```python
from PIL import Image
import cmath
import math


def function(z: complex):
    return cmath.exp(3 * cmath.log(z, cmath.e)) - 1


def function_der(z: complex):
    return 3 * cmath.exp(2 * cmath.log(z, cmath.e))


def get_roots():
    return complex(1, 0), complex(-0.5, math.sqrt(3) / 2), complex(-0.5, -math.sqrt(3) / 2)


def get_colors():
    return (153, 0, 153), (0, 204, 0), (51, 51, 255), (0, 0, 0)


# declare constants
x_min, x_max = -5, 5
y_min, y_max = -5, 5
it_max = 25
it_eps = 10 ** -3
roots = get_roots()
colors = get_colors()
# image size
window_size = 1000

image = Image.new("RGB", (window_size, window_size))
print(100 * "|", end='')
print()
cnt = 0
for y in range(window_size):
    cnt += 1
    for x in range(window_size):
        colored = False

        y_coord = y * (y_max - y_min) / (window_size - 1) + y_min
        x_coord = x * (x_max - x_min) / (window_size - 1) + x_min
        z = complex(x_coord, y_coord)

        for it in range(it_max):
            z -= function(z) / function_der(z)

            for r, root in enumerate(roots):
                diff = z - root
                if abs(diff.real) < it_eps and abs(diff.imag) < it_eps:
                    image.putpixel((x, y), colors[r])
                    colored = True
        if not colored:
            image.putpixel((x, y), colors[-1])
    if cnt % (window_size // 100) == 0:
        print("|", end='')

image.show()
```
